```python
# Author - Ritvik Khanna 
# Date - 04/05/18 
# Version - 1.0

import pandas as pd

# import os
# print(os.listdir("../input"))
# Load dataset
df = pd.read_csv("car_evaluation.csv", names = ["buying","maint", "doors", "persons", "lug_boot","safety","class"])
print(df.head(10))
df.describe()
```

```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','red','green','blue','orange','white','brown']
df["buying"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
df["maint"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
df[ "doors"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
df[ "persons"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
df["lug_boot"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
df["safety"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
df["class"].value_counts().plot(kind='pie',title='luggage boot',colors=colors)
plt.show()
```
